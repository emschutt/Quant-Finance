```python
import pandas as pd
import numpy as np
import yfinance as yf
import fredapi
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
from datetime import datetime
```

```python
start = "2015-01-01"
end = "2025-01-01"

# DGS10 is in percent (e.g., 4.25), daily business days
rf = pdr.DataReader("DGS10", "fred", start, end).dropna()

# convert to decimal for pricing models (e.g., 0.0425)
rf["DGS10_decimal"] = rf["DGS10"] / 100.0

print(rf.tail())
latest_rf = rf["DGS10_decimal"].iloc[-1]
print("Latest 10Y risk-free rate (decimal):", latest_rf)
```
